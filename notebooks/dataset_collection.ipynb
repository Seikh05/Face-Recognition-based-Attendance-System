# Face Recognition System: Dataset Collection

This notebook will help you create a dataset of face images. It opens your webcam, uses Haar Cascades to detect your face, and saves a cropped image of your face to the `dataset` folder every 0.5 seconds.

### Instructions:
1. Run the entire cell below.
2. Enter your name when prompted.
3. Look at the webcam while it captures images.
4. Press `q` if you want to abort the capture early. The capture will automatically stop after 30 images.

```python
import cv2
import os
import time

# 1. Define the person's name and create their dataset directory
person_name = input("Enter the name of the person: ")
dataset_dir = os.path.join("..", "dataset", person_name)

# Create the directory if it does not exist
if not os.path.exists(dataset_dir):
    os.makedirs(dataset_dir)
    print(f"Created directory for {person_name} at: {dataset_dir}")
else:
    print(f"Directory already exists at: {dataset_dir}")

# 2. Load the Haar Cascade for face detection
# OpenCV provides pre-trained cascades; we use the default frontal face cascade
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# 3. Open the webcam using OpenCV
# The argument 0 refers to the default camera
cap = cv2.VideoCapture(0)

# 4. Set parameters for image capture
capture_interval = 0.5  # Time in seconds between captures
max_images = 200        # Maximum number of images we want to save
count = 0               # Counter for captured images
last_capture_time = time.time()  # Track the time since last capture

print(f"Starting to capture images for {person_name}. Please look at the camera.")

while cap.isOpened() and count < max_images:
    # Read a frame from the webcam
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab a frame from the webcam.")
        break
        
    # Convert the frame to grayscale for the face detection algorithm
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # 5. Detect faces in the frame
    # scaleFactor=1.3 compensates for varying distance to camera
    # minNeighbors=5 specifies how many neighbors each candidate rectangle should have
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5, minSize=(30, 30))
    
    # 6. Process each detected face
    for (x, y, w, h) in faces:
        # Draw a blue rectangle around the detected face
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
        
        # Check if 0.5 seconds have passed since the last capture
        current_time = time.time()
        if current_time - last_capture_time >= capture_interval:
            # 7. Extract the region of interest (the face) and save it
            face_roi = frame[y:y+h, x:x+w]
            
            # Format the filename as person_name_0000.jpg, person_name_0001.jpg, etc.
            img_path = os.path.join(dataset_dir, f"{person_name}_{count:04d}.jpg")
            
            # Save the face image to the dataset directory
            cv2.imwrite(img_path, face_roi)
            print(f"Captured {img_path}")
            
            count += 1
            last_capture_time = current_time
            
            # Optional: Stop checking other faces in THIS frame to avoid saving multiple
            # people as the same person in crowded backgrounds.
            break 
            
    # Display the live video feed with rectangles
    cv2.imshow('Face Data Collection - Press Q to stop', frame)
    
    # Break the loop early if 'q' is pressed on the keyboard
    if cv2.waitKey(1) & 0xFF == ord('q'):
        print("Collection interrupted by user.")
        break

# 8. Clean up resources
# Release the webcam
cap.release()
# Close all OpenCV windows
cv2.destroyAllWindows()
print(f"Collection completed successfully. Saved {count} images in {dataset_dir}.")
```
